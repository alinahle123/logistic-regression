```python
def mesh(X, h = 0.02):
    x_min , x_max = X.iloc[:, 0]. min() - .5, X.iloc[:, 0]. max() + .5
    y_min , y_max = X.iloc[:, 1]. min() - .5, X.iloc[:, 1]. max() + .5
    xx , yy = np.meshgrid(np.arange(x_min , x_max , h), np.arange(y_min , y_max , h))
    return np.c_[xx.ravel (), yy.ravel ()], xx , yy
```

```python
def accuracy(y, y_hat):
    accuracy = np.sum(y == y_hat) / len(y)
    return accuracy
```

```python
def logreg_cost(X, y, w):
    z = X@w
    cost=0
    z=np.asarray(z)
    for i in range (len(y)):
        cost=(cost+y[i]*z[i])+np.log(1+np.exp(-z[i]))
    return cost
```

```python
def logreg_grad_desc(X, y):
    ones=np.ones(X.shape[0])
    X['ones']=ones
    w = [0 for i in range(X.shape[1])]
    
    wcopy=w.copy()
    nb_iter = 1000 # to tune
    pas = 0.00007 # to tune
    cost2=0
    cost_now=0
    #beta = 0.9 # for backtracking if needed - to tune
    for i in range(nb_iter):
        z = X @ wcopy
        z=z.reset_index(drop=True)
        grad=np.dot(np.transpose(X),(y-((np.exp(-z)//(1+np.exp(-z))))))
        for j in range(len(w)):
            wcopy[j]=wcopy[j]-pas*grad[j]
        cost2=cost_now
        cost_now=logreg_cost(X,y,wcopy)
    
        if(cost_now<0):
            if(-cost_now<cost2):
                w=wcopy.copy()
            break
        else:
            w=wcopy.copy()
    return w,logreg_cost(X,y,w)

w,cost=logreg_grad_desc(X,y)
print(w,cost)
```

### les fonctions ci-dessus sont :
### 1-accuracy : qui nous donnera la précision de notre prédiction.
### 2-logreg_cost : qui calculera le coût et nous l'utiliserons pour améliorer les poids et nous calculons le coût en utilisant la formule que nous avons vue en classe.
### 3-logreg_grad_desc : nous donne le meilleur w basé sur le moindre coût ( où pas=0.00007 et nb_iter=1000 basé sur des essais ce sont les paramètres qui nous ont donné les meilleurs résultats) et la fonction calcule les poids et le gradient basé sur les formules que nous avons vu dans la classe

```python
def predict(X,w):
    Xcopy=X.copy()
    ones=np.ones(X.shape[0])
    Xcopy['ones']=ones
    pred=(1/(1+np.exp(-(Xcopy@w))))
    return pred
pred=predict(X,w)
```

### the predict function predicts the labels of the X dataset based on the weights that we have got from the previous function

```python
x1 = X['Points']
x2 = X['Y']
x3 = [min(X['Points']), max(X['Points'])]
m = -w[0]/w[1]
c = -w[2]/w[1]
x4 = m*np.asarray(x3) + c
plt.scatter(x1[y==-1], x2[y==-1], c='r')
plt.scatter(x1[y==1], x2[y==1], c='b')
plt.plot(x3,x4,'-y')
```

```python
pred_class = [-1 if i > 0.5 else 1 for i in pred]
accuracy(y,pred_class)
```

### as we can see this our accuracy:96.1%

### dans l'ensemble de données sur le cancer, nous n'avons pas besoin de la précision, mais plutôt d'un taux de vrais positifs plus élevé, c'est-à-dire de prédire correctement les cas qui ont un cancer, nous allons donc diviser l'ensemble de données, former nos modèles, les tester sur l'ensemble de données de test et voir le taux de vrais positifs de chaque algorithme.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
Cancer1=datasets.load_breast_cancer()
Cancer = pd.DataFrame(Cancer1.data, columns=Cancer1.feature_names)
Cancer_y=Cancer1.target
Cancer=(Cancer-Cancer.mean())/Cancer.std()
# Cancer_y=np.where(Cancer_y==0, -1,Cancer_y)
# print(Cancer_y)
Cancer.head()
Cancer_Train,Cancer_Test,y_Train,y_Test=train_test_split(Cancer,Cancer_y, test_size=0.3, random_state=42)
```

```python
clf = LogisticRegression(random_state=0,C=0.05).fit(Cancer_Train,y_Train)

```

```python
pred_scikit=clf.predict(Cancer_Test)
```

```python
w,cost=logreg_grad_desc(Cancer_Train,y_Train)
```

```python
pred=predict(Cancer_Test,w)
pred_class = [0 if i > 0.5 else 1 for i in pred]
```

```python
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    y_hat=np.asarray(y_hat)
    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1

    return(TP*100/108, FP*100/63, TN*100/63, FN*100/108)
```

### cette fonction calcule le Vrai positif (prédit correctement le cancer), le Vrai négatif (prédit correctement qu'il n'y a pas de cancer), le Faux positif (prédit qu'il y a un cancer alors qu'il n'y en a pas) et le Faux négatif (le pire) (prédit qu'il n'y a pas de cancer alors qu'il y en a).

```python
print(perf_measure(y_Test,pred_class))
```

### Comme nous pouvons le voir, notre algorithme a un taux de vrais positifs de 100% et c'est ce que nous visons, mais on a un taux de Faux positifs de 12%

```python
print(perf_measure(y_Test,pred_scikit))
```

### aussi le scikit_learn a 100% de vrais positifs, mais il peut être amélioré (il a des faux positifs) 3.1%.

```python
def confusion_matrix_scorer(clf, X, y):
      y_pred = clf.predict(X)
      cm = confusion_matrix(y, y_pred)
      return cm[0,0]
```

```python
model = LogisticRegression()
solvers = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
penalty = ['l2','l1']
c_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
max_iter=[100,200,500,1000]
tol=[0.1,0.001,0.001,0.0001]
grid = dict(solver=solvers,penalty=penalty,C=c_values,max_iter=max_iter,tol=tol)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring=confusion_matrix_scorer,error_score=0)
grid_result = grid_search.fit(Cancer_Train, y_Train)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

### Ici, nous avons procédé au réglage des hyperparamètres afin d'obtenir le meilleur modèle pour cet ensemble de données. Une chose que j'ai changé est le paramètre de score (le score sur lequel nous choisissons le meilleur modèle) et j'ai choisi le taux de vrais positifs comme paramètre de score.
